##I. Пакет Matplotlib

###1a. Прочитайте текст из User’s Guide “Pyplot tutorial” (текст лекции, стр. 2–21) или https://matplotlib.org/stable/tutorials/introductory/pyplot.html#sphx-glr-tutorialsintroductory-pyplot-py и выполните команды из него

График зависимости x от y. X начинается с нуля, y -- с 1, потому что так задали мы сами.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([1, 2, 3, 4])
plt.ylabel('Ось y')
plt.xlabel('Ось x')
plt.show()
```

Так можно задать координаты x и y.

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel('Ось y')
plt.xlabel('Ось x')
plt.show()
```

Третий аргумент (необязательный) определяет, как отображать график. В этом примере r -- это красный, o -- это круг.

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

В этом примере третий аргумент у функции plt.plot обозначает, что линия графика должна быть непрерывной и зелёного цвета.

Функция axis принимает список [xmin, xmax, ymin, ymax], определяя размеры график.

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], '-g')
plt.axis([0, 5, 0, 18])
plt.ylabel('Ось y')
plt.xlabel('Ось x')
plt.show()
```

Вместо списков можно использовать массивы numpy.

```python
import numpy as np
# равномерно распределенное время с интервалом в 200 мс
t = np.arange(0., 5., 0.2)
# красные черточки, синие квадраты и зеленые треугольники
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

Некоторые команды, которые имеют параметр data, позволяют передавать данные в виде словаря. Например, команды numpy.recarray или pandas.DataFrame. plt.scatter -- это график разброса.

```python
data = {'a': np.arange(50),
 'c': np.random.randint(0, 50, 50),
 'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
plt.figure(1) # the first figure
plt.subplot(211) # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212) # the second subplot in the first figure
plt.plot([4, 5, 6])
plt.figure(2) # a second figure
plt.plot([4, 5, 6]) # creates a subplot() by default
plt.figure(1) # figure 1 current; subplot(212) still current
plt.subplot(211) # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

Работа с текстом. Помимо названия осей xlabel и ylabel можно добавлять текст внутри графика с помощью функции plt.text, в которой задаются координаты и текст.

```python
mu, sigma = 100, 20
x = mu + sigma * np.random.randn(10000)
# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .02, r'$\mu=100,\ \sigma=20$')
plt.axis([40, 160, 0, 0.025])
plt.grid(True)
plt.show()
```

Подписи создаются с помощью функции plt.annotate, там задаются 2 точки: расположение аннотируемого объекта xy и расположение текста xytext. Оба этих аргумента задаются кортежами вида (x, y).

```python
ax = plt.subplot()
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
 arrowprops=dict(facecolor='black', width=0.01, shrink=0.05) )
plt.ylim(-2, 2)
plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)
# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))
# plot with various axes scales
plt.figure()
# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('Линейная шкала', fontsize = 10)
plt.xticks(fontsize = 7)
plt.grid(True)
# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('Логарифмическая шкала', fontsize = 10)
plt.xticks(fontsize = 7)
plt.grid(True)
# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('вблизи нуля линейный, а в остальной – логарифмический', fontsize = 7)
plt.grid(True)
# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('Функция логит', fontsize = 8)
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
 wspace=0.35)
```

### 2. Выберите один пример из набора https://matplotlib.org/stable/gallery/index.html. Разберите его и выполните команды из него (рассказать/пояснить код: что делают команды, что задают параметры команд).

Grouped bar chart with labels. Сгруппированная столбчатая диаграмма с метками.

```python
species = ("Адэли", "Чинстрап", "Генту")
penguin_means = {                         # словарь, состоящий из attribute и measurement
    'Толщина ключа': (18.35, 18.43, 14.98), # толщина клюва
    'Длина клюва': (38.79, 48.83, 47.50), # длина ключа
    'Длина ластов': (189.95, 195.82, 217.19), # длина ластов
}

x = np.arange(len(species))  # расположение меток, одномерный массив с равномерно разнесенными значениями внутри заданного интервала
width = 0.25  # ширина столбцов
multiplier = 0

fig, ax = plt.subplots(layout='constrained') # параметр constrained позволяет вписать график в фигуру

for attribute, measurement in penguin_means.items():
    offset = width * multiplier # сдвиг -- это ширина столбца на множитель
    rects = ax.bar(x + offset, measurement, width, label=attribute) # рисует столбцы
    ax.bar_label(rects, padding=3) # подписи к столбцу. Padding показывает, на сколько точек от конца столбца написать метку
    multiplier += 1

# Добавление текста, меток, легенды. Изменяем параметры осей (axis).
ax.set_ylabel('Высота (мм)') # устанавливаем название оси y
ax.set_title('Свойства пингвина по видам') # название графика
ax.set_xticks(x + width, species) # установка меток для оси x
ax.legend(loc='upper left', ncol=2) # расположили легенду в верхнем левом углу, шириной с две колонки
ax.set_ylim(0, 250) # левый и правый пределы для оси y на графике

plt.show()
```

Основы построения графиков: команда plot

```python
ts = pd.Series(np.random.randn(1000), index = pd.date_range("1/1/2000", periods = 1000))
ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index = ts.index, columns = list("ABCD")).cumsum()
df.plot()
```

Можно рисовать график зависимости одной величины от другой (x от y).

```python
df2 = pd.DataFrame(np.random.randn(1000, 2), columns = ["B", "C"]).cumsum()
df2["A"] = pd.Series(list(range(len(df))))
df2.plot(x="A", y="B")
```

```python
plt.figure()
df.iloc[5].plot(kind="bar")
```

# III. Пакет Seaborn

### 4. Прочитайте текст “An introduction to seaborn” https://seaborn.pydata.org/tutorial/introduction.html и выполните команды из него.

```python
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
tips.head()
```

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```

```python
# a linear regression mode с помощью lmplot
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
sns.catplot(data=tips, kind="swarm", x="day", y="total_bill", hue="smoker")
```

```python
sns.catplot(data=tips, kind="violin", x="day", y="total_bill", hue="smoker", split=True)
```

```python
sns.catplot(data=tips, kind="bar", x="day", y="total_bill", hue="smoker")
```

```python
penguins = sns.load_dataset("penguins")
sns.set_theme(style="ticks", font_scale=1.25)
g = sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g",
    palette="crest", marker="x", s=100,
)
g.set_axis_labels("Bill length (mm)", "Bill depth (mm)", labelpad=10)
g.legend.set_title("Body mass (g)")
g.figure.set_size_inches(6.5, 4.5)
g.ax.margins(.15)
g.despine(trim=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid") # стиль -- белая сетка

# Задаётся фигура matplotlib
f, ax = plt.subplots(figsize=(12, 15))

# Загружается датасет
crashes = sns.load_dataset("car_crashes").sort_values("total", ascending=False)

# График общего количества аварий
sns.set_color_codes("pastel")
sns.barplot(x="total", y="abbrev", data=crashes, # x, y берутся из датасета
            label="Общее количество аварий", color="b")

# График аварий, связанных с алкоголем
sns.set_color_codes("muted")
sns.barplot(x="alcohol", y="abbrev", data=crashes, # x, y берутся из датасета
            label="Связанные с алкоголем", color="b")

# Добавляется легенда и метки у осей
ax.legend(ncol=2, loc="lower right", frameon=True) # ncol -- количество колонок. frameon -- нужна ли рамка.
ax.set(xlim=(0, 24), ylabel="", # xlim начальную и конечную координаты
       xlabel="Автоаварии на миллиард миль")
sns.despine(left=True, bottom=True) # If True, remove that spine.
```
