### Задача 1: Поиск URL-адресов

Напишите функцию, которая принимает текстовую строку и находит все URL-адреса в этой строке с помощью регулярных выражений. Например, если в строке есть текст "Посетите наш сайт: https://www.example.com", функция должна вернуть "https://www.example.com".

Будем считать, что URL-адрес состоит из нескольких частей:

**<схема>://<хост>/<URL‐путь>**

* Протокол: http или https
* Хост - доменное имя сайта без "/" (может включать www)
* <URL‐путь> - уточняющая информация о месте нахождения ресурса

Более подробное описание можно посмотреть в [Википедии](https://ru.wikipedia.org/wiki/URL). Там представлена полная схема URL-адреса, но в данной задаче мы будем пользоваться упрощенной, которая описана выше

```python
import re
```

```python
url = re.compile(r"https?://[a-zA-Z0-9\./\-_]*")
url.findall(text)
```

### Задача 3: Подсчет частоты слов

Например, если входная строка "Я люблю кошек, и кошки любят меня", функция должна вернуть словарь: {"Я": 2, "любить": 2, "кошка": 2, "и": 1}.

*Указание: предварительно лемматизируйте текст*

Загрузка MyStem для лемматизации текста

```python
!wget http://download.cdn.yandex.net/mystem/mystem-3.0-linux3.1-64bit.tar.gz
!tar -xvf mystem-3.0-linux3.1-64bit.tar.gz
!cp mystem /bin
!pip install git+https://github.com/nlpub/pymystem3 #установка последней версии

!git clone https://github.com/Mottl/ru_punkt #загрузка библиотеки

from pymystem3 import Mystem
```

```python
myst = Mystem()
```

Функция для лемматизации текста. Необязательная, если использовать сразу функцию count_lemmas().

```python
def to_lemmatize(s):
  lemmatized = ''.join(myst.lemmatize(s)).strip()
  return lemmatized
```

Функция для подсчёта слов в лемматизированном тексте. Текст предварительно лемматизировать не нужно, потому что лемматизация происходит внутри функции.

```python
def count_lemmas(s):
  lemmatized = ''.join(myst.lemmatize(s)).lower().strip()
  
  words = lemmatized.split()

  count_words = dict()
  for word in words:
    if word in count_words:
      count_words[word] += 1
    else:
      count_words[word] = 1

  return count_words
```

```python
t = "ЖЕ что же же чТО такое было же должно было случается"
```

```python
count_lemmas(t)
```

### Задача 5: Поиск списка фраз в леммтизированном тексте с извлечением из исходного

Поиск при этом должен осуществляться в лемматизированном тексте. Результат должен быть отсортирован от самого раннего вхождения к самому позднему.

Пример: 

* Текст: "Сегодня так много солнечных лучей и совсем нет долждя. Дождь будет завтра"

* Список фраз для поиска ["солнечный луч", "дождь", "сегодня"]

Желаемый результат: ["сегодня", "солнечных лучей", "дождя", "дождь"]


**Пояснение**: хотя "сегодня" находится в конце в списка для поиска, это слово встречается раньше всех в тексте. Значит, оно должно быть в начале результирующего списка.
